```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#导入DataFrame格式数据
iris_sns=sns.load_dataset("iris")
#fig为画布ax为显示的轴对象限制大小为8*6英寸
fig,ax=plt.subplots(figsize=(8,6))
#绘制直方图,指定了x轴的数据  可以使用stat表示纵轴类型:概率概率密度,bins表示区间数量
#sns绘制模式不容,是基于plt的状态,所以不需要提前赋值
#seaborn默认使用pltgca()获取当前ax进行绘图,当然也可自己传入指定ax
sns.histplot(data=iris_sns,x="sepal_length",
            binwidth=0.2,ax=ax,hue='species')
#取均值mean表示计算这一列的均值,这时候x只有一个值
ax.axvline(x=iris_sns.sepal_length.mean(),
           color='r',ls='--')
ax.axvline(x=iris_sns.sepal_length.mean() + iris_sns.sepal_length.std(),
           color='b',ls='--')
ax.axvline(x=iris_sns.sepal_length.mean() - iris_sns.sepal_length.std(),
           color='b',ls='--')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#导入DataFrame格式数据
iris_sns=sns.load_dataset("iris")
#fig为画布ax为显示的轴对象限制大小为8*6英寸
fig,ax=plt.subplots(figsize=(8,6))
#绘制直方图,指定了x轴的数据  可以使用stat表示纵轴类型:概率概率密度,bins表示区间数量
#sns绘制模式不容,是基于plt的状态,所以不需要提前赋值
#seaborn默认使用pltgca()获取当前ax进行绘图,当然也可自己传入指定ax
sns.histplot(data=iris_sns,x="sepal_length",
            binwidth=0.2,ax=ax,hue='species',
             #element默认为普通条形,这里step表示空心
             #stat控制y轴的统计方式,默认为计数
            element="step",stat='density')
#取均值mean表示计算这一列的均值,这时候x只有一个值
ax.axvline(x=iris_sns.sepal_length.mean(),
           color='r',ls='--')
ax.axvline(x=iris_sns.sepal_length.mean() + iris_sns.sepal_length.std(),
           color='b',ls='--')
ax.axvline(x=iris_sns.sepal_length.mean() - iris_sns.sepal_length.std(),
           color='b',ls='--')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#导入数据
iris_sns=sns.load_dataset("iris")

#创建指定画布大小的fig
fig,ax = plt.subplots(figsize=(8,6))
#使用高斯核密度绘图
sns.kdeplot(data=iris_sns,x='sepal_length',
            #调整平滑程度   fill为填充曲线下方的区域
           bw_adjust=0.3,fill=True)
#绘制毛毯图
sns.rugplot(data=iris_sns,x='sepal_length')

#不同颜色分开进行绘制
fig,ax = plt.subplots(figsize=(8,6))
sns.kdeplot(data=iris_sns,x='sepal_length',hue='species',
           bw_adjust=0.3,fill=True)
sns.rugplot(data=iris_sns,x='sepal_length',hue='species')


fig,ax = plt.subplots(figsize=(8,6))
sns.kdeplot(data=iris_sns,x='sepal_length',hue='species',
            #这里让密度曲线堆叠显示
           multiple='stack',bw_adjust=0.5)


fig,ax = plt.subplots(figsize=(8,6))
sns.kdeplot(data=iris_sns,x='sepal_length',
            hue='species',bw_adjust=0.5,multiple='fill')#密度归一化
```

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.stripplot(data=iris_sns,x='sepal_length',y='species',
              #hue使用petal_length给数据点进行着色 使用palette的颜色方案进行着色
              hue='petal_length',palette='RdYlBu_r',ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.swarmplot(data=iris_sns,x="sepal_length",ax=ax)

fig,ax=plt.subplots(figsize=(8,6))
sns.swarmplot(data=iris_sns,x="sepal_length",y='species',
              hue='species',ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(8,2))
sns.boxplot(data=iris_sns,x='sepal_length',ax=ax)

fig,ax=plt.subplots(figsize=(8,2))
sns.boxplot(data=iris_sns,x='sepal_length',
            y='species',ax=ax)
```

```python
#小提琴图
fig,ax=plt.subplots(figsize=(8,2))
sns.violinplot(data=iris_sns,x='sepal_length',ax=ax)
#小提琴图+分类
fig,ax=plt.subplots(figsize=(8,4))
sns.violinplot(data=iris_sns,x='sepal_length',
               y='species',ax=ax)

#蜂群图+小提琴图
#catplot本身会创建一个新的fig,无需创建 可以使用kind指定图形类型
sns.catplot(data=iris_sns,x='sepal_length',y='species',
            kind='violin',color='.9',inner=None)#inner这里让图片不显示内部数据
sns.swarmplot(data=iris_sns,x='sepal_length',
              y='species',size=3)
```
