```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#导入DataFrame格式数据
iris_sns=sns.load_dataset("iris")
#fig为画布ax为显示的轴对象限制大小为8*6英寸
fig,ax=plt.subplots(figsize=(8,6))
#绘制直方图,指定了x轴的数据  可以使用stat表示纵轴类型:概率概率密度,bins表示区间数量
#sns绘制模式不容,是基于plt的状态,所以不需要提前赋值
#seaborn默认使用pltgca()获取当前ax进行绘图,当然也可自己传入指定ax
sns.histplot(data=iris_sns,x="sepal_length",
            binwidth=0.2,ax=ax)
#取均值mean表示计算这一列的均值,这时候x只有一个值
ax.axvline(x=iris_sns.sepal_length.mean(),
           color='r',ls='--')
```
