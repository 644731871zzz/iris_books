```python
import seaborn as sns

iris_sns=sns.load_dataset('iris')

sns.scatterplot(data=iris_sns,x='sepal_length',y='petal_width')
```

```python
sns.displot(data=iris_sns,x='sepal_length',y='petal_width',cbar=True)
```

```python
sns.displot(data=iris_sns,x='sepal_length',y='petal_width',kind='kde')
```
