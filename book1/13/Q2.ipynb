```python
import numpy as np
import matplotlib.pyplot as plt

x1_array=np.linspace(-3,3,100)
x2_array=np.linspace(-3,3,100)

xx1,xx2=np.meshgrid(x1_array,x2_array)
ff=xx2*np.exp(-xx1**2-xx2**2)

fig,ax=plt.subplots(figsize=(8,8))
ax.contour(xx1,xx2,ff,levels=10,colors='black',linewidths=0.5)
ax.contourf(xx1,xx2,ff,levels=50,cmap='coolwarm')

plt.show()
```
