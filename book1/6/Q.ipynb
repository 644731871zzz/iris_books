```python
#Q2
import math

r=2
circualr_area=math.pi*r*2
print(f'circualr_area:{circualr_area:.2f}')
```

```python
#Q3
import matplotlib.pyplot as plt

num=100
#创建了函数的起始值和最终值,并求出了公差
x_start=-3
x_end=3
x_step=(x_end-x_start)/(num-1)

#这里将所有函数中的x进行配置 并且创建了需要后续用的y的数列集合
x_array=[x_start+i*x_step for i in range(num)]
y_array=[]

#计算y的值并使用for配置到y_array
for x in x_array:
    y=math.exp(x)
    y_array.append(y)

#绘制图像
plt.plot(x_array,y_array)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
#Q4
#这里计算高斯函数添加到新的数列
y_gauss_array=[]
for x in x_array:
    y=math.exp(-x**2)
    y_gauss_array.append(y)
plt.plot(x_array,y_gauss_array)
plt.grid(True)
```

```python
#Q5 质地均匀的筛子
import random
import statistics
#投掷次数
num_flips=3000
#计算结果,6个面的概率一样
results=[random.choices([1,2,3,4,5,6])[0] for _ in range(num_flips)]
#计算均值并存储
running_means=[statistics.mean(results[0:idx+1]) for idx in range(num_flips)]

#绘制均值图像
plt.plot(range(1,num_flips+1),running_means)
plt.axhline(3.5,color='r')
plt.xlabel("Number of flips")
plt.ylabel("Running means")
plt.grid(True)
plt.ylim(1,6)
plt.show()

#修改概率后的新的均值变化图
#新的结果
new_results=[random.choices([1,2,3,4,5,6],[0.2,0.16,0.16,0.16,0.16,0.16])[0] for _ in range(num_flips)]
new_running_means=[statistics.mean(new_results[0:idx+1]) for idx in range(num_flips)]
plt.plot(range(1,num_flips+1),new_running_means)
plt.axhline(3.5,color='r')
plt.xlabel("Number of flips")
plt.ylabel("Running means")
plt.grid(True)
plt.ylim(1,6)
plt.show()
```
