```python
import numpy as np
import matplotlib.pyplot as plt

#生成横轴数据
x_array=np.linspace(0,2*np.pi,100)
sin_y=np.sin(x_array)
cos_y=np.cos(x_array)

#设置图片大小
fig,ax=plt.subplots(figsize=(8,6))

#绘制正弦和余弦曲线
ax.plot(x_array,sin_y,
        label='sin',color='b',linewidth=2)
ax.plot(x_array,cos_y,
        label='cos',color='r',linewidth=2)

ax.set_title('Sine and cosine functions')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

#添加图例
ax.legend()

#设置横轴和纵轴坐标范围
ax.set_xlim(0,2*np.pi)
ax.set_ylim(-1.5,1.5)

#设置横轴标签和刻度标签
x_ticks=np.arange(0,2*np.pi+np.pi/2,np.pi/2)
x_ticklabels=[r'$0$',r'$\frac{\pi}{2}$',
              r'$\pi$',r'$\frac{3\pi}{2}$',
              r'$2\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)

#横纵轴采用相同scale
ax.set_aspect('equal')
plt.grid()
#将图片储存成svg格式
plt.savefig('正弦余弦函数曲线.svg',format='svg')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2*np.pi,100)
y_sin=np.sin(x)
y_cos=np.cos(x)

fig,(ax1,ax2)=plt.subplots(1,2,
                           figsize=(10,4),
                           sharey=True)

ax1.plot(x,y_sin,color='blue')
ax1.set_title('Sine function')
ax1.set_xlabel('x')
#设置y轴标签 第二个rotation表示旋转角度 这里代表将图片旋转到水平方向(水平显示) ha表示指定y标签对其方式
ax1.set_ylabel('f(x)',
               rotation='horizontal',
               ha='right')
ax1.set_xlim(0,2*np.pi)
ax1.set_ylim(-1.5,1.5)
#为什么进制是一个单位pi
x_ticks=np.arange(0,2*np.pi+0.001,np.pi)
#指定x轴上的 标签文本 列表.都表示字符串使用r和$表示可以插入特殊字符
x_ticklabels=[r'$0$',r'$\pi$',r'$2\pi$']
ax1.set_xticks(x_ticks)
ax1.set_xticklabels(x_ticklabels)
#添加网格
ax1.grid(True)
#配置x与y的比例相等 equal表示等比例
ax1.set_aspect('equal')

ax2.plot(x,y_cos,color='red')
ax2.set_title('Cosine function')
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)',
               rotation='horizontal',
               ha='right')
ax2.set_xlim(0,2*np.pi)
ax2.set_ylim(-1.5,1.5)
ax2.set_xticks(x_ticks)
ax2.set_xticklabels(x_ticklabels)
ax2.grid(True)
ax2.set_aspect('equal')

#自动调整两幅图片比例
plt.tight_layout()

plt.show()
```

```python
import plotly.express as px
import numpy as np

#生成横轴数据
x=np.linspace(0,2*np.pi,100)

y_sin=np.sin(x)
y_cos=np.cos(x)

fig=px.line(x=x,y=[y_sin,y_cos],labels={'y':'f(x)','x':'x'})
#修改图例
fig.data[0].name='Sine'
fig.data[1].name='Cosine'

fig.show()
```

```python
import plotly.express as px
import numpy as np
import pandas as pd

#生成横轴数据
x=np.linspace(0,2*np.pi,100)

#生成正弦余弦曲线的数据
y_sin=np.sin(x)
y_cos=np.cos(x)

#生成pandas数据帧
df=pd.DataFrame({'x':x,'Sine':y_sin,'Cosine':y_cos})

#创建图标
fig=px.line(df,x='x',y=['Sine','Cosine'],
            labels={'value':'f(x)','X':'x'})

fig.show()
```
