```python
#填充两个曲线之间的区域,叫做填充之间或者填充区域
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
x=np.linspace(0,8,16)
#len函数确定数量为16个 0.0-0.5这个区间是左闭右开 所以数字中没有0.5存在
y1=3+4*x/8 + np.random.uniform(0.0,0.5,len(x))
y2=1+2*x/8+np.random.uniform(0.0,0.5,len(x))

fig,ax=plt.subplots()

#alpha是透明度,范围为0-1,linewidth表示没有区域边缘的额外线条,默认没有边缘,所以设置宽度也不会影响
ax.fill_between(x,y1,y2,alpha=0.5,linewidth=0)
#先绘制底层然后上层,所以没有被
ax.plot(x,(y1+y2)/2 , linewidth=2)

ax.set(xlim=(0,8),xticks=np.arange(1,8),
       ylim=(0,8),yticks=np.arange(1,8))

plt.show()
```

```python
#图像处理叫做imshow imshow(Z)将二维数组(简单矩阵)作为像素质按照颜色显示 像是热力图
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

#linspace生成[-3,3]的16个等差数字
#meshgrid是输入两个方形,一个x一个y 然后输出X和Y的矩阵,对应了范围坐标轴内x*y的格式的矩阵,X映射矩阵中对应位置的x值Y为y的值
X,Y=np.meshgrid(np.linspace(-3,3,16),np.linspace(-3,3,16))
#形成数学曲面,并使用高斯分布
Z=(1-X/2+X**5+Y**3)*np.exp(-X**2-Y**2)

fig,ax=plt.subplots()

#因为Z为矩阵,直接显示
ax.imshow(Z,origin='lower')

plt.show()
```

```python
#绘制等高线
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

X,Y=np.meshgrid(np.linspace(-3,3,256),np.linspace(-3,3,256))
Z=(1-X/2+X**5+Y**3)*np.exp(-X**2-Y**2)
#控制等高线的范围表示有7条线,从最大到最小
levels=np.linspace(np.min(Z),np.max(Z),7)

fig,ax=plt.subplots()

ax.contour(X,Y,Z,levels=levels)

plt.show()
```

```python
#箭头形状的图  矢量场图
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

x=np.linspace(-4,4,6)
y=np.linspace(-4,4,6)
X,Y=np.meshgrid(x,y)
#U与V表示X方向与V方向上的分量(水平与垂直方向的长度)
U=X+Y
V=Y-X

fig,ax=plt.subplots()

#angles为角度计算方式表示xy的计算方式 scale units表示以'xy'表示坐标缩放 然后是箭头长度(越大越短,作用于矢量场,值是比例缩放的倒数,表示积分值几 表示分母的字)与宽度(越大越粗)的配置
ax.quiver(X,Y,U,V,color="C0",angles='xy',
          scale_units='xy',scale=5,width=0.015)
ax.set(xlim=(-5,5),ylim=(-5,5))

plt.show()
```

```python
#直方图 histogram histo表示柱子 gram表示记录
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

np.random.seed(1)
#生成高斯分布的随机数200各 均值为0 1.5为标准差 4+是对所有生成的数加上4
#均值决定曲线中心位置,标准差决定曲线宽度,这里大部分落在-3,3之间 然后加上4表示向右移动了4个单位,数据将会集中在4附近
x=4+np.random.normal(0,1.5,200)

fig,ax=plt.subplots()

#直方图不需要y因为由hist统计然后显示
#hist分成多个区间,然后统计每个区间的数据个数(频数)
#bins表示有8个柱子.hist会计算几个柱子再计算柱子中的数值个数
#仅有200数据点,所以仅从最小值到最大值这个区间均匀分为8份以左闭右开的规则
#然后统计每一个区间内的x个数来绘制y的图
ax.hist(x,bins=8,linewidth=0.5,edgecolor="white")

#range生成相等间隔整数 linspace生成刻度,非连续数字
ax.set(xlim=(0,8),xticks=np.arange(1,8),
       ylim=(0,56),yticks=np.linspace(0,56,9))

plt.show()
```

```python
#三维曲面图
import matplotlib.pyplot as plt
import numpy as np
#导入颜色映射模块
from matplotlib import cm

plt.style.use('_mpl-gallery')

X=np.arange(-5,5,0.25)
Y=np.arange(-5,5,0.25)
#创建网格坐标 将XY向上向右复制平铺,使得可以描述某点坐标
X,Y=np.meshgrid(X,Y)
#点到原点之间距离
R=np.sqrt(X**2+Y**2)
#用sin表示一个周期函数
Z=np.sin(R)

fig,ax=plt.subplots(subplot_kw={"projection":"3d"})
#vmin颜色映射的最小值 每一个值对应了一个颜色,如果使用Z的较小的值,表示其他值一定比这个大
ax.plot_surface(X,Y,Z,vmin=Z.min()*2,cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
#三维三角网格曲面图
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

plt.style.use('_mpl-gallery')

n_radii=8
n_angles=36

radii=np.linspace(0.125,1.0,n_radii)
#生成2-2pi的等间距数组,一共有_angles个数据 不包括终点
#后面的[]表示增加一个新维度,生成二维数组
angles=np.linspace(0,2*np.pi,n_angles,endpoint=False)[...,np.newaxis]

#flatten将二维数组压缩成一维数组
x=np.append(0,(radii*np.cos(angles)).flatten())
y=np.append(0,(radii*np.sin(angles)).flatten())
z=np.sin(-x*y)

fig,ax=plt.subplots(subplot_kw={"projection":"3d"})

ax.plot_trisurf(x,y,z,vmin=z.min()*2,cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])


plt.show()
```

```python
#三维体素图 体素类似二维的像素
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

#返回多维数组索引值,标志xyz坐标
x,y,z=np.indices((8,8,8))

#cube表示布尔数值形成立方体
cube1=(x<3)&(y<3)&(z<3)
cube2=(x>=5)&(y>=5)&(z>=5)

#|表示或
voxelarray=cube1|cube2

fig,ax=plt.subplots(subplot_kw={"projection":"3d"})

ax.voxels(voxelarray,edgecolor='k')

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])


plt.show()
```

```python
#三维柱状图
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

x=[1,1,2,2]
y=[1,2,1,2]
z=[0,0,0,0]
#代表了立方体的长宽高 相等是为了创建正方体的底
#one函数返回一个与x相同的数组
dx=np.ones_like(x)*0.5
dy=np.ones_like(x)*0.5
dz=[2,3,1,4]

fig,ax=plt.subplots(subplot_kw={"projection":"3d"})

ax.bar3d(x,y,z,dx,dy,dz)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])


plt.show()
```

```python
#三维曲线图
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

n=100
#x是从0,1增加
xs=np.linspace(0,1,n)
#y是将每个数都是原来的6*pi倍的数使用sin进行处理
ys=np.sin(xs*6*np.pi)
zs=np.cos(xs*6*np.pi)

fig,ax=plt.subplots(subplot_kw={"projection":"3d"})

ax.plot(xs,ys,zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])


plt.show()
```

```python
#三维stem图 显示数据点到水平面的图
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

n=20
x=np.sin(np.linspace(0,2*np.pi,n))
y=np.cos(np.linspace(0,2*np.pi,n))
z=np.linspace(0,1,n)

fig,ax=plt.subplots(subplot_kw={"projection":"3d"})

ax.stem(x,y,z)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])


plt.show()
```
